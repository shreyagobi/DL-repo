```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),                 # Convert to tensor [0,1]
    transforms.Normalize((0.5,), (0.5,))   # Normalize to mean=0, std=1
])
```

```python
train_dataset = datasets.FashionMNIST(
    root="./data",
    train=True,
    transform=transform,
    download=True
)

test_dataset = datasets.FashionMNIST(
    root="./data",
    train=False,
    transform=transform,
    download=True
)
```

```python
batch_size = 64

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
class DeepFFNN(nn.Module):
    def __init__(self):
        super(DeepFFNN, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(28 * 28, 512),
            nn.ReLU(),

            nn.Linear(512, 256),
            nn.ReLU(),

            nn.Linear(256, 128),
            nn.ReLU(),

            nn.Linear(128, 10)  # 10 classes
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten image
        return self.model(x)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DeepFFNN().to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    running_loss = 0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # ----- Forward Pass -----
        outputs = model(images)
        loss = criterion(outputs, labels)

        # ----- Backward Pass -----
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    accuracy = 100 * correct / total
    print(f"Epoch [{epoch+1}/{num_epochs}], "
          f"Loss: {running_loss:.4f}, Accuracy: {accuracy:.2f}%")
```

```python
import torch
import torch.nn as nn

# ✅ DEFINE FUNCTION INLINE (NO DEPENDENCIES)
def visualize_activations(model, image):
    activations = []
    x = image.unsqueeze(0)  # [1, 1, 28, 28]

    for layer in model:
        x = layer(x)
        activations.append(x.detach().cpu())

    return activations
```

```python
images, _ = next(iter(test_loader))
image = images[0].to(device)

# Redefine visualize_activations with the fix to flatten the input
def visualize_activations(model, image):
    activations = []
    x = image.unsqueeze(0)  # [1, 1, 28, 28]
    x = x.view(x.size(0), -1) # Flatten the input tensor here to [1, 784]

    for layer in model:
        # Some layers (like activation functions) don't change the shape in a way that
        # makes them incompatible with subsequent linear layers, but nn.Linear expects 2D.
        x = layer(x)
        activations.append(x.detach().cpu())

    return activations

activations = visualize_activations(model, image)

for i, act in enumerate(activations):
    print(f"Layer {i+1} activation shape: {act.shape}")
```
